Step 3: Feature Engineering

```python
merged_data_all['nutrient_density']=(merged_data_all['protein (g)']+merged_data_all['fiber (g)'])/(merged_data_all['carbs (g)']+merged_data_all['fat (g)'])
```

```python
merged_data_all.groupby('meal_type')['nutrient_density'].mean()
```

```python
merged_data_all.columns
```

```python
merged_data_all['total_carbs_a_day'] = merged_data_all.groupby(['user_id','day'])['carbs (g)'].transform('sum')
```

```python
merged_data_all['total_carbs_a_day'].describe()
```

```python
merged_data_all['blood_sugar_impact_score'] = merged_data_all['blood_sugar_level'] * merged_data_all['nutrient_density']
```

```python
merged_data_all.groupby('meal_type')['blood_sugar_impact_score'].mean()
```

```python
merged_data_all['nutrient_per_meal'] = merged_data_all['protein (g)'] + merged_data_all['fiber (g)']
merged_data_all['total_nutrient_intake'] = merged_data_all.groupby(['user_id','day'])['nutrient_per_meal'].transform('sum')
```

```python
threshold = merged_data_all['total_nutrient_intake'].quantile(0.8)
merged_data_all['heavy_meal_day'] = np.where(merged_data_all['total_nutrient_intake'] > threshold, 'Yes', 'No')
```
